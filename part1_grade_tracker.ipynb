```python
# ============================================================
# Assignment - Part 1: Python Basics & Control Flow
# Theme: Student Grade Tracker
# Total Marks: 25
# ============================================================

# ============================================================
# TASK 1 - Data Parsing & Profile Cleaning (5 marks)
# ============================================================

print("=" * 50)
print("       TASK 1 - Data Parsing & Profile Cleaning")
print("=" * 50)

# Given raw student data (names messy, roll as string, marks as single string)
raw_students = [
    {"name": " ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",      "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": " Priya Nair ",     "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",      "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",   "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

students = []  # This will hold all cleaned student records

for s in raw_students:
    # Clean name: remove extra spaces + convert to Title Case (e.g. "ayesha sharma" -> "Ayesha Sharma")
    name = s["name"].strip().title()

    # Convert roll from string to integer
    roll = int(s["roll"])

    # Split marks_str by ", " and convert each number to integer
    marks = [int(m) for m in s["marks_str"].split(", ")]

    # Validate name: every word must contain only alphabetic characters
    is_valid = all(word.isalpha() for word in name.split())
    validity_label = "/ Valid name" if is_valid else "x Invalid name"

    # Store cleaned data
    cleaned = {"name": name, "roll": roll, "marks": marks, "valid": is_valid}
    students.append(cleaned)

    # Print formatted profile card using f-strings
    print("=" * 30)
    print(f"Student : {name}  {validity_label}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("=" * 30)
    print()

# Find roll 103 and print in ALL CAPS and lowercase
for s in students:
    if s["roll"] == 103:
        print(f"Roll 103 - ALL CAPS  : {s['name'].upper()}")
        print(f"Roll 103 - lowercase : {s['name'].lower()}")


# ============================================================
# TASK 2 - Marks Analysis Using Loops & Conditionals (8 marks)
# ============================================================

print("\n" + "=" * 50)
print("   TASK 2 - Marks Analysis Using Loops & Conditionals")
print("=" * 50)

student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# Function to get grade based on marks range
def get_grade(mark):
    if mark >= 90:   return "A+"
    elif mark >= 80: return "A"
    elif mark >= 70: return "B"
    elif mark >= 60: return "C"
    else:            return "F"

# 1. Print each subject with marks and grade
print(f"\nStudent: {student_name}")
print(f"{'Subject':<12} {'Marks':>6} {'Grade':>6}")
print("-" * 28)
for subject, mark in zip(subjects, marks):
    print(f"{subject:<12} {mark:>6} {get_grade(mark):>6}")

# 2. Calculate and print totals
total   = sum(marks)
average = round(total / len(marks), 2)
highest = subjects[marks.index(max(marks))]
lowest  = subjects[marks.index(min(marks))]

print(f"\nTotal marks             : {total}")
print(f"Average marks           : {average}")
print(f"Highest scoring subject : {highest} ({max(marks)})")
print(f"Lowest scoring subject  : {lowest} ({min(marks)})")

# 3. Simulate marks-entry system using a while loop
# NOTE: Using pre-defined test inputs to demonstrate logic without blocking.
# For real interactive use, replace the entries list with:
#   subject_input = input("Enter subject name (or 'done' to exit): ")
#   mark_input    = input(f"Enter marks for {subject_input}: ")

print("\n--- Simulating New Subject Entry (while loop) ---")
entries = [("Biology", "82"), ("Hindi", "74"), ("done", "")]  # simulated user inputs
new_subjects = []
new_marks    = []

i = 0
while i < len(entries):
    subject_input, mark_input = entries[i]
    print(f"Enter subject name: {subject_input}")

    if subject_input.lower() == "done":
        print("Stopping entry.")
        break

    print(f"Enter marks for {subject_input}: {mark_input}")
    try:
        mark_val = int(mark_input)
        if 0 <= mark_val <= 100:
            new_subjects.append(subject_input)
            new_marks.append(mark_val)
            print(f"  Added: {subject_input} -> {mark_val}")
        else:
            print(f"  Warning: Marks must be 0-100. '{mark_val}' is invalid, not added.")
    except ValueError:
        print(f"  Warning: '{mark_input}' is not a valid number, not added.")
    i += 1

all_marks    = marks + new_marks
updated_avg  = round(sum(all_marks) / len(all_marks), 2)
print(f"\n{len(new_subjects)} new subject(s) added.")
print(f"Updated average across all {len(all_marks)} subjects: {updated_avg}")


# ============================================================
# TASK 3 - Class Performance Summary (7 marks)
# ============================================================

print("\n" + "=" * 50)
print("        TASK 3 - Class Performance Summary")
print("=" * 50)

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma",   [55, 68, 49, 72, 61]),
    ("Priya Nair",    [91, 85, 88, 94, 79]),
    ("Karan Mehta",   [40, 55, 38, 62, 50]),
    ("Sneha Pillai",  [75, 80, 70, 68, 85]),
]

results = []

# 1. Compute average and status for each student
for name, marks_list in class_data:
    avg    = round(sum(marks_list) / len(marks_list), 2)
    status = "Pass" if avg >= 60 else "Fail"
    results.append({"name": name, "avg": avg, "status": status})

# 2. Print formatted class report table
print(f"\n{'Name':<20} | {'Average':>7} | Status")
print("-" * 42)
for r in results:
    print(f"{r['name']:<20} | {r['avg']:>7} | {r['status']}")

# 3. Summary stats
passed    = [r for r in results if r["status"] == "Pass"]
failed    = [r for r in results if r["status"] == "Fail"]
topper    = max(results, key=lambda x: x["avg"])
class_avg = round(sum(r["avg"] for r in results) / len(results), 2)

print(f"\nStudents passed  : {len(passed)}")
print(f"Students failed  : {len(failed)}")
print(f"Class topper     : {topper['name']} ({topper['avg']})")
print(f"Class average    : {class_avg}")


# ============================================================
# TASK 4 - String Manipulation Utility (5 marks)
# ============================================================

print("\n" + "=" * 50)
print("       TASK 4 - String Manipulation Utility")
print("=" * 50)

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and web development. developers love python for its clean and readable syntax. python makes programming accessible to everyone.  "

# Step 1: Strip leading/trailing whitespace
clean_essay = essay.strip()
print(f"\nStep 1 - Stripped:\n{clean_essay}\n")

# Step 2: Convert to Title Case
title_essay = clean_essay.title()
print(f"Step 2 - Title Case:\n{title_essay}\n")

# Step 3: Count "python" (case-insensitive)
# clean_essay is already lowercase so we can count directly
count_python = clean_essay.count("python")
print(f"Step 3 - Count of 'python' in essay: {count_python}\n")

# Step 4: Replace all "python" with "Python"
replaced_essay = clean_essay.replace("python", "Python")
print(f"Step 4 - After replacing 'python':\n{replaced_essay}\n")

# Step 5: Split into individual sentences on ". " (period + space)
sentences = clean_essay.split(". ")
print("Step 5 - Sentences after splitting on '. ':")
for s in sentences:
    print(f"   '{s}'")

# Step 6: Print each sentence numbered, add "." if it doesn't already end with one
print("\nStep 6 - Numbered sentences:")
for i, sentence in enumerate(sentences, 1):
    sentence = sentence.strip()
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")

```
